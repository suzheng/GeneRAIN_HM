```python
def get_rank_and_simi_of_one_exp(rank_simi_results_dicts, exp, gene_pairs_label):
    rank_simi = rank_simi_results_dicts[exp][gene_pairs_label]
    rank_dict = rank_simi[0]
    simi_dict = rank_simi[1]
    return rank_dict, simi_dict
```

```python
def calculate_top_k_stats(value_list, k_values=[1, 5, 10, 20, 50, 100, 200, 500, 1000, 2000]):
    """
    Calculate the top k hits statistics for a given list of values.

    Parameters:
    value_list (list): A list of rank values.
    k_values (list): A list of k values for which to calculate the top k hits.

    Returns:
    dict: A dictionary containing the count and fraction of hits for each top k value.
    """
    value_array = np.array(value_list)
    results = {}

    total_values = len(value_list)
    for k in k_values:
        hits = np.sum(value_array <= k)
        results[f"top{k}"] = hits
        results[f"top{k}_frac"] = hits / total_values*100

    return results
```

```python
# Get dataframe for subsampling the orthologous gene pairs supervision
exp = njson.exp
tot = 10
res_dict_list = []
for i in range(1, tot+1):
    one_res_rank, one_res_simi = get_rank_and_simi_of_one_exp(after_muse_rank_simi_results_dicts,
                                           f"{exp}.subsampled_{i}of{tot}",
                                           njson.h2m_one2one
                                         )
    rank_list = list(one_res_rank.values())
    res = {"frac": i/10, **calculate_top_k_stats(rank_list)}
    res_dict_list.append(res)
subsampled_df = pd.DataFrame(res_dict_list)
```

```python
# visualize performance over training epochs
exp = njson.exp
res_dict_list = []
for epoch in [1, 3, 5, 10, 15, 20, 30, 40]:
    one_res_rank, one_res_simi = get_rank_and_simi_of_one_exp(after_muse_rank_simi_results_dicts,
                                                              f"{exp}.epoch{epoch}",
                                                              njson.h2m_one2one
                                                             )
    rank_list = list(one_res_rank.values())
    res = {"epoch": epoch, **calculate_top_k_stats(rank_list)}
    res_dict_list.append(res)
epoch_df = pd.DataFrame(res_dict_list)
```

```python
def get_cos_simi_of_full_spv_for_gene_pairs(exp, h2m_one2one):
    m_param_json_file = f"/g/data/yr31/zs2131/tasks/2023/RNA_expr_net/anal_mouse/training/{exp.split('_')[0]}/{exp}.param_config.json"
    os.environ['PARAM_JSON_FILE'] = m_param_json_file
    
    params = params_class()
    gene2idx, _ = get_gene2idx(params.GENE_EMB_NAME)
    gene_to_ori_case_dict = get_gene_to_ori_case_dict(gene2idx)
    exp = os.path.basename(m_param_json_file).split(".")[0]
    out_dir = config.proj_path + f"/results/anal/{exp}/"
    dir_without_uid = f"{out_dir}/{exp}.epoch{epoch}.full_supervised/"
    
    muse_result_dir = get_muse_result_uid_dir(dir_without_uid)
    hm_emb, hm_gene_names, hm_token_types = load_muse_result(muse_result_dir, gene_to_ori_case_dict=gene_to_ori_case_dict)
    similarity_matrix = cosine_similarity(hm_emb)
    id2idx = {name: idx for idx, name in enumerate(hm_gene_names)}
    cos_simi = [get_similarity(h_gene, config.gene_prefix+v, similarity_matrix, id2idx) for h_gene, v in h2m_one2one.items()]
    return cos_simi
```

```python
def calculate_correlations(x_values, y_values):
    # Calculating Spearman's correlation
    spearman_corr, spearman_p = spearmanr(x_values, y_values)

    # Calculating Pearson's correlation
    pearson_corr, pearson_p = pearsonr(x_values, y_values)

    return {
        "Spearman's correlation": spearman_corr,
        "Spearman's p-value": spearman_p,
        "Pearson's correlation": pearson_corr,
        "Pearson's p-value": pearson_p
    }
```
